```python
import numpy as np
import matplotlib.pyplot as plt

import astropy.units as u
from astropy import coordinates

import galenv

%matplotlib inline
```

```python
ge = galenv.Galenv()
```

## By name

```python
ge.conedv_byname("ACO 13", 10, 5000, show=True) # objname, tangential_dist, dv
```

### By name but no redshift in NED

```python
ge.conedv_byname("2MASX J03514350+3841573", 10, 5000, show=True)
```

### By name but user input velocity < 3000

```python
ge.conedv_byname("2MASX J03514350+3841573", 2, 5000, show=True)
```

## By position & velocity

```python
co = coordinates.SkyCoord(ra=80.7416, dec=-36.45857, unit=(u.deg, u.deg))  # "PKS J0522-3627"
```

```python
ge.conedv_byposvel(co, 16952, 0.1, 5000, show=True) # coord, vel, tangential_dist, dv
```

### By position but user input of velocity is less than 3000 km/s

```python
ge.conedv_byposvel(co, 2500, 2, 5000, show=True) # coord, vel, tangential_dist, dv
```
