```python
import numpy as np
import matplotlib.pyplot as plt

#from astropy.cosmology import FlatLambdaCDM
from astropy.cosmology import Planck15 # FlatLambdaCDM with parameters from Planck Collab 2015, Paper XIII
import astropy.units as u
from astroquery.ned import Ned
import matplotlib.gridspec as gridspec

import galenv

%matplotlib inline
```

```python
objname = "PKS J0522-3627"
```

```python
galenv.run(objname, 0.1, 1000, show=True)
```

### Plot all data (even objects with no redshift/velocity information)

```python
z, v0, ra, dec = galenv.queryobject_by_name(objname)

result = Ned.query_region(objname, radius=1*u.arcmin)
```

```python
len(result)
```

```python
galenv.plot_env(ra, dec, 0.017, result, v0, 5000, show=True, savefig=True, imgname="PKS.png")
```
